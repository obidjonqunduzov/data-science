```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(np.random.randn(3,4), index=list('abc'), columns=['Olma','Nok','Anor','Gilos'])
```

```python
df
```

```python
np.abs(df) # df ning absoluit qiymatini hisoblaydi
```

```python
np.round(df) # df qiymatlarini yaxlitlaydi
```

```python
df.mean() # df ning o'rtacha qiymatini hisoblaydi
```

```python
df.apply(np.sum, axis='columns') # df ni ustun bo'yicha hisoblaydi
```

```python
df # asl df o'zgarmagan holatda
```

```python
df_2 = pd.DataFrame(np.arange(1,17).reshape(4,4), index=list('abcd'), columns=['Olma','Nok','Anor','Gilos'])
```

```python
df_2
```

```python
df.apply(lambda x: x.max() - x.min()) # df ning max qiymatlaridan min qiymatlarini ayirmasini hisoblaydi
```

```python
def foo(x):
  return pd.Series([x.min(), x.max()], index=['min', 'max'])
```

```python
df.apply(foo) # df ni min va max qiymatlarini hisoblaydi
```

```python
df_2.applymap(lambda x: x*x) # df_2 ning har bir elementini kvadratga ko'taradi
```
