```python
import pandas as pd
```

```python
import numpy as np
```

```python
uz = {2023:35.5, 2022:34, 2021:33.42, 2020:32.75, 2018:31.95, 2015:28.3}
kz = {2023:21.4, 2022:20.8, 2021:19.35, 2020:18.56, 2018:16.8, 2015:14.32}
tj = {2023:11.2, 2022:10.75, 2021:9.93, 2020:8.7, 2018:7.14, 2015:6}
kg = {2023:7.5, 2022:6.89, 2021:5.4, 2020:4.8, 2018:3.7, 2015:2.5}

df = pd.DataFrame({
    'Uzbekistan':uz,
    'Kazakhstan':kz,
    'Tajikistan':tj,
    'Kirgizistan':kg
})
df
```

```python
df.loc[[2018]] # loc metodi yordamida 2018 yilgi davlatlar aholisi sonini chiqarib beradi
```

```python
df.loc[[2020,2021]] # ikkita yilni chiqarib beradi
```

```python
df.loc[[2023,2015], ['Tajikistan','Uzbekistan']] # yil va davlatlar nomi oraqali chiqarib beradi
```

```python
df
```

```python
df.iloc[[0,1,4]] # iloc bu qator va ustunlarning indexsi bo'yicha qaytaradi
```

```python
df.iloc[[2],[1,2]] # 2 qator va 0 dan 3 gacha bo'lgan ustunni qaytaradi
```

```python
df.iloc[[0,1,2,3],[1,2,3]]
```

```python
df.loc[2023:,'Uzbekistan'] # 2023 dan kiyingi barcha qiymatlarni Series turi bo'yicha qaytaradi
```
