```python
import pandas as pd
```

```python
data_user_russia = {
    'year':[1990, 1995, 2000, 2005, 2010, 2015, 2020, 2023],
    'born':[147665081, 1363806,1266800,	1457376, 1788948, 1940579, 1436514, 1398253],
    'temp':[1.892, 1.337, 1.195, 1.294, 1.567, 1.777, 1.579, 1.504]
}
```

```python
data_info = pd.DataFrame(data_user_russia)
data_info
```

```python
data_info.shape # ustun va qatorlari
```

```python
data_info = pd.DataFrame(data_user_russia, columns=['year','temp','born']) # ustunlar joyini o'zgartiradi
data_info
```
