```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/usa_cars.csv', index_col=0)
df.head()
```

```python
df['reating'] = df['price'].rank() # ro'yxatdagi avtomobillarni price(narxi) bo'yicha yangi reating ustunida reytinglab chiqdi
```

```python
df.sort_values(by='price') # shu reytinglangan ustunimizni ketma-ket tartiblab chiqdi
```

```python
df.sort_values('mileage') # mileage(bosib o'tgan) yo'li boyicha retnglarini tartiblab chiqaradi
```

```python
# .rank() method parametri yordamida reyting o'lchash usulini tanlash mumkin (min, max, average, first, dense)
df['reating'] = df['year'].rank(method='min') # avtomobillarni yili boyicha retingladi
df.sort_values('year') # va shu reytingni min(eng kam) qiymatlaridan boshlab ketma-ket tartiblab chiqaradi
```

```python
df['reating'] = df['price'].rank(method='max') # avtomobillarni yili boyicha retingladi
df.sort_values('price') # va shu reytingni max(eng k'p) qiymatlaridan boshlab ketma-ket tartiblab chiqaradi
```

```python
df['reating'] = df['price'].rank(method='average') # avtomobillarni yili boyicha retingladi
df.sort_values('price') # va shu reytingni average(o'rtacha) qiymatlaridan boshlab ketma-ket tartiblab chiqaradi
```
