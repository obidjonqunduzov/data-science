```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/usa_cars.csv', index_col=0)
df.head()
```

```python
df.sort_index()
```

```python
df.sort_index(axis=1) # ustun nomi bo'yicha tartiblash
```

```python
df.sort_index(axis=1, ascending=False) # ustunlarni teskarisiga tartiblash
```

```python
df.sort_values(by='price', ascending=False) # price ustunini eng katta qiymati bo'yicha tartibladi
```

```python
df.sort_values(by=['model','brand'], ) # model va brand ustunlarini ketma-ket tartibladi
```

```python
df.sort_values(by=['mileage','price'], ascending=False) # mileage(yurgan yo'li) va price(narxi) bo'yicha tartibladek
```
