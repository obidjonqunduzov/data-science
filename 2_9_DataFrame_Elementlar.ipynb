```python
import pandas as pd
```

```python
import numpy as np
```

```python
uz = {2023:35.5, 2022:34, 2021:33.42, 2020:32.75, 2018:31.95, 2015:28.3}
kz = {2023:21.4, 2022:20.8, 2021:19.35, 2020:18.56, 2018:16.8, 2015:14.32}
tj = {2023:11.2, 2022:10.75, 2021:9.93, 2020:8.7, 2018:7.14, 2015:6}
kg = {2023:7.5, 2022:6.89, 2021:5.4, 2020:4.8, 2018:3.7, 2015:2.5}

df = pd.DataFrame({
    'Uzbekistan':uz,
    'Kazakhstan':kz,
    'Tajikistan':tj,
    'Kirgizistan':kg
})
df
```

```python
df['Uzbekistan'] # Series malumotlarini qaytaradi
```

```python
df['Uzbekistan'].to_numpy() # numpydagi array qiymatini qaytaradi
```

```python
df[['Uzbekistan']] # DataFrame malumotlarini qaytaradi
```

```python
df[['Tajikistan', 'Kirgizistan']]
```

```python
df[3:] # 3chi elementlardan kiyingilarini qaytaradi
```

```python
df>10 # aholisi soni 10mln dan katta bo'lgan davlatlarni qaytaradi
```

```python
df[['Uzbekistan']][df['Uzbekistan']>32] # Uzbekistan aholisini  32mln ko'p bo'lgan yillarni DataFrame ko'rinishida qaytaradi
```

```python
df['Tajikistan']>8 # Tajikistanni aholisini 8mln oshgan yillarini numpy ko'rinishida qaytaradi
```

```python
df2 = df.transpose(copy=True) # Elementlar ustun va qatorlarini almashtiradi
df2
```

```python
df2[df2[2018]>3] # 2018 yil aholisi 3mln ko'p davlatlarni qaytaradi
```

```python
df2[df2[2015]<8] # 2015 aholisi 8mln kam davlatlarni qaytaradi
```
